#### DATA PREPROCESSING 

- Dropping redundant columns

```python
cdata.columns
```

```python
cdata.drop(["Unnamed: 0","Photo","Flag","Club Logo"],axis=1,inplace=True)
cdata.columns
```

```python
cdata.head()
```

- Checking for duplicates

```python
cdata.duplicated().value_counts()
```

There exists 52 duplicate records which should be dropped

```python
cdata.drop_duplicates(inplace=True)
cdata.duplicated().value_counts()
```

Here we have dropped all the duplicate values

- Preprocessing individual columns

```python
cdata.columns
```

Value column : We will convert it into float data type

```python
cdata.Value.head()
```

Now few values are in million euros and few are in thousand euros , so we need to process values accordingly 

€105M --> 105000000.0 & €105K --> 105000

```python
cdata.Value.replace("€","",regex=True,inplace=True)
cdata.Value.replace("M","*1000000",regex=True,inplace=True)
cdata.Value.replace("K","*1000",regex=True,inplace=True)
cdata.Value=cdata.Value.apply(eval)
```

```python
cdata.Value.head()
```

Here we have successfully converted the calue column in required format

Wage columns : We will convert it into float data type

```python
cdata.Wage.head()
```

All values are in thousand euros

```python
cdata.Wage.replace("€","",regex=True,inplace=True)
cdata.Wage.replace("K","*1000",regex=True,inplace=True)
cdata.Wage=cdata.Wage.apply(eval)
```

```python
cdata.Wage.head()
```

Here we have successfully converted the Wage column in required format

Acceleration column : It contain values like '73+10', '71+6', '77+13','12' so we will evaluate them into a whole number 

```python
cdata.Acceleration.unique()[-5:]
```

```python
cdata.Acceleration=cdata.Acceleration.apply(lambda x:eval(x))
```

```python
cdata.Acceleration.unique()[-5:]
```

```python
cdata.Acceleration.unique()
```

Here we have successfully converted the Acceleration column in required format

Just like acceleration column , all the columns from acceleration to volleys contain similar type of not required values like '73+10', '71+6', '77+13' , so we will evaluate them into a whole number all together instead of repeating the process for individual columns

```python
cdata.loc[:,"Aggression":"Volleys"]=cdata.loc[:,"Aggression":"Volleys"].applymap(lambda x:eval(x))
```

```python
cdata.Dribbling.head(3),cdata.Volleys.head(3)
```

Here we have successfully converted the values of columns from Aggression to volleys into required format 

```python
cdata.info()
```

Thus coverted all the records in desired format and data type

- Handling null values

```python
nullv = cdata.isnull().sum()
nullv[nullv!=0]
```

As per above series we can see column club consists of 248 null values and rest consist of 2024 null values , we will fill the club null values with "No data" since those values for club are not known and rest null values with 0

```python
cdata.Club=cdata.Club.fillna("No Data")
cdata=cdata.fillna(0)
cdata.info()
```

Here we have processed all the null values

```python
d = pd.DataFrame(cdata.groupby("Club")["Preferred Positions"].value_counts())
d.rename(columns={"Preferred Positions":"Count"},inplace=True)
d.reset_index(inplace=True)

plt.figure(figsize=(16,4))
plt.subplot(1,3,1)
b = d[d.Club=="FC Barcelona"].nlargest(5,"Count")
plt.plot(b["Preferred Positions"],b.Count,color="k",marker="o",linestyle="--",markersize=5)
plt.ylabel("Count")
plt.xlabel("FC Barcelona")
plt.grid(linewidth=0.5)

plt.subplot(1,3,2)
j = d[d.Club=="Juventus"].nlargest(5,"Count")
plt.plot(j["Preferred Positions"],j.Count,color="k",marker="o",linestyle="--",markersize=5)
plt.xlabel("Juventus")
plt.grid(linewidth=0.5)

plt.subplot(1,3,3)
Rm = d[d.Club=="Real Madrid CF"].nlargest(5,"Count")
plt.plot(Rm["Preferred Positions"],Rm.Count,color="k",marker="o",linestyle="--",markersize=5)
plt.xlabel("Real Madrid CF")
plt.grid(linewidth=0.5)
```

Inferences;
- Comparison between top 3 teams on the basis of overall values
- FC Barcelona focuses more on goal keeper location followed by center back and line backer location 
- Juventus focuses more on center back location followed by central defensive midfielder and goal keeper location
- Real Madrid FC focuses equally on center back, central defensive midfielder and goal keeper location
